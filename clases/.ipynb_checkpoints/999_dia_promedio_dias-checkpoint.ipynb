```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import numpy as np
```

```python
f = '../data/002_processed/esolmet_2018.parquet'
tmx = pd.read_parquet(f)
tmx['To'] = tmx['To'].replace(0,np.nan)
tmx
```

```python
promedio = tmx.To.groupby(by=tmx.index.strftime('%H:%M')).mean()
promedio.index = pd.to_datetime(promedio.index, format='%H:%M')

desviacion = tmx.To.groupby(by=tmx.index.strftime('%H:%M')).std()
desviacion.index = pd.to_datetime(desviacion.index, format='%H:%M')

maximos = tmx.To.groupby(by=tmx.index.strftime('%H:%M')).max()
maximos.index = pd.to_datetime(maximos.index, format='%H:%M')

minimos = tmx.To.groupby(by=tmx.index.strftime('%H:%M')).min()
minimos.index = pd.to_datetime(minimos.index, format='%H:%M')
```

```python
promedio
```

```python
fig, ax = plt.subplots(figsize=(12,4))

ax.plot(promedio)
ax.fill_between(promedio.index, promedio - desviacion, promedio + desviacion, alpha = 0.2)
ax.plot(maximos,'r.', alpha = 0.2)
ax.plot(minimos,'r.', alpha = 0.2)

ax.set_ylabel('Temperatura [oC]')
ax.set_xlabel('Hora [HH:MM]')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax.xaxis.set_major_locator(mdates.HourLocator(interval=2))

plt.show()
```

```python
To = tmx['To']
g = To.groupby(To.index.normalize())

fig, ax = plt.subplots(figsize=(12,4))
for _, day in g:
    x = day.index.map(lambda t: t.replace(year=2000, month=1, day=1))
    ax.plot(x, day.values, alpha=0.05, lw=1,c='gray')

Tomax = tmx['To'].loc['2018-05-30'] 
h = Tomax.groupby(Tomax.index.normalize())
for _, day in h:
    x = day.index.map(lambda t: t.replace(year=2000, month=1, day=1))
    ax.plot(x, day.values, alpha=1, lw=2,c='k')
    


ax.xaxis.set_major_locator(mdates.HourLocator(interval=3))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
ax.margins(x=0); plt.tight_layout()
ax.set_xlabel('tiempo [HH:MM]')
ax.set_ylabel('Temperatura [$^oC$]')
```

```python

theta = 2*np.pi*((tmx.index.hour*60 + tmx.index.minute)/1440.0)
r = tmx["To"].values

ax = plt.subplot(111, projection='polar')
ax.scatter(theta, r, s=1, alpha=0.01)
ax.set_xticks(np.linspace(0, 2*np.pi, 8, endpoint=False))
ax.set_xticklabels(['00:00','03:00','06:00','09:00','12:00','15:00','18:00','21:00'])
plt.tight_layout()
```

```python

serie = tmx.To.sort_index()
dias = serie.groupby(serie.index.normalize())  # una serie por fecha

fig = plt.figure(figsize=(6,6))
ax = plt.subplot(111, projection='polar')
ax.set_theta_zero_location('N')   # 00:00 arriba
ax.set_theta_direction(-1)        # sentido horario

for _, day in dias:
    theta = 2*np.pi*((day.index.hour*60 + day.index.minute)/1440.0)
    # (opcional) cerrar la curva del dÃ­a:
    theta = np.r_[theta, 2*np.pi]
    r = np.r_[day.values, day.values[0]]
    ax.plot(theta, r, lw=0.8, alpha=0.15)

ax.set_xticks(np.linspace(0, 2*np.pi, 8, endpoint=False))
ax.set_xticklabels(['00:00','03:00','06:00','09:00','12:00','15:00','18:00','21:00'])
plt.tight_layout()
plt.show()
```
