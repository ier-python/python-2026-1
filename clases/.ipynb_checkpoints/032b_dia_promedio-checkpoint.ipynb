```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = "../data/002_processed/esolmet_2018.parquet"
tmx = pd.read_parquet(f)
tmx
```

```python
tmx.groupby(by=[tmx.index.month, tmx.index.hour])[["To"]].mean()
```

## Como hago :

1. Un dia promedio de cada mes incluyendo las horas y minutos desde el inicio
2. Un dia promedio con tooodos los datos

```python
# import numpy as np
resumen_mensual = (
    tmx.groupby(pd.Grouper(freq="ME"))
       .agg(
           Ig_sum  = ("Ig", "sum"),          # energía/irradiación acumulada (si aplica)
           To_mean = ("To", "mean"),         # temp. media mensual
           ws_max  = ("ws", "max"),          # racha máxima
           uv_p95  = ("uv", lambda s: s.quantile(0.95))  # percentil 95 de UV
       )
       .rename_axis("Mes")
)
resumen_mensual
```
